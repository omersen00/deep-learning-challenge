```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df = application_df.drop(columns = ['EIN', 'SPECIAL_CONSIDERATIONS', 'STATUS'])
application_df
```

```python
application_df.nunique()
```

```python
ask_amt = application_df['ASK_AMT'].value_counts()
ask_amt
```

```python
ask_amt_replace = list(ask_amt[ask_amt < 50].index)

for app in ask_amt_replace:
    application_df['ASK_AMT'] = application_df['ASK_AMT'].replace(app,"Other")

application_df['ASK_AMT'].value_counts()
```

```python
application_type = application_df['APPLICATION_TYPE'].value_counts()
application_type
```

```python
application_type_replace = list(application_type[application_type < 50].index)

for app in application_type_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

application_df['APPLICATION_TYPE'].value_counts()
```

```python
incame_amt = application_df['INCOME_AMT'].value_counts()
incame_amt
```

```python
classifications_to_replace = list(incame_amt[incame_amt < 50].index)

for cls in classifications_to_replace:
    application_df['INCOME_AMT'] = application_df['INCOME_AMT'].replace(cls,"Other")


application_df['INCOME_AMT'].value_counts()
```

```python
app_dummies = pd.get_dummies(application_df)
app_dummies
```

```python
X = app_dummies.drop(['IS_SUCCESSFUL'], axis= 1)
y = app_dummies['IS_SUCCESSFUL']

X_train, X_test, y_train, y_test = train_test_split(X, y,)
```

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train.iloc[0])
hidden_nodes_layer1 =  80
hidden_nodes_layer2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
